```python
# Import libraries
import pandas as pd
import numpy as np
from datetime import datetime

#  Step 1: Drop 'PROFESSION' if it exists
if 'PROFESSION' in df.columns:
    df.drop(columns=['PROFESSION'], inplace=True)

#  Step 2: Fill missing values

# Numerical: fill with 0
if 'NOMBRE_ENFANT' in df.columns:
    df['NOMBRE_ENFANT'] = df['NOMBRE_ENFANT'].fillna(0)

# Categorical: fill with mode or "unknown"
for col in ['MARITAL_STATUS', 'SEXE', 'COUNTRY', 'VILLE', 'RESIDENCE']:
    if col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna('unknown')
        else:
            df[col] = df[col].fillna(df[col].mode()[0])

#  Step 3: Process AGE from DATE_NAISSANCE

if 'DATE_NAISSANCE' in df.columns:
    # Convert to datetime safely
    df['DATE_NAISSANCE'] = pd.to_datetime(df['DATE_NAISSANCE'], errors='coerce')

    # Keep only realistic birthdates
    df = df[(df['DATE_NAISSANCE'] >= '1900-01-01') & 
            (df['DATE_NAISSANCE'] <= pd.to_datetime("today")) | 
            df['DATE_NAISSANCE'].isnull()]

    # Compute age
    today = pd.to_datetime("today")
    df['AGE'] = (today - df['DATE_NAISSANCE']).dt.days // 365

    # Fill missing age with median
    df['AGE'] = df['AGE'].fillna(df['AGE'].median())

    # Drop date of birth column
    df.drop(columns=['DATE_NAISSANCE'], inplace=True)

#  Final check
print("Cleaned dataset info:")
display(df.info())

print("\nAGE column stats:")
display(df['AGE'].describe())

print("\nRemaining missing values:")
display(df.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

# Separate features and target
X = df.drop(columns=["credit_obtenu"])
y = df["credit_obtenu"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# detect column types
cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
num_cols = X.select_dtypes(include=["number"]).columns.tolist()

print("Categorical:", cat_cols)
print("Numeric:", num_cols)

# numeric pipeline: impute median
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
])

# categorical pipeline:
# 1) impute missing with most frequent
# 2) force everything to string (avoid float+str mix)
# 3) one-hot encode to numeric
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("to_str", FunctionTransformer(lambda data: data.astype(str))),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])

preprocess = ColumnTransformer(
    transformers=[
        ("cat", categorical_transformer, cat_cols),
        ("num", numeric_transformer, num_cols),
    ]
)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 1) Take a manageable sample of the training set for SMOTE
#    (here: max 300k rows – adjust if your RAM allows more)
max_smote_rows = 300_000

if len(X_train) > max_smote_rows:
    X_smote, _, y_smote, _ = train_test_split(
        X_train, y_train,
        train_size=max_smote_rows,
        stratify=y_train,
        random_state=42
    )
else:
    X_smote = X_train.copy()
    y_smote = y_train.copy()

print("SMOTE will be applied on:", X_smote.shape)

# 2) Fit preprocessing on this smaller subset and transform it
X_smote_enc = preprocess.fit_transform(X_smote)
print("Encoded SMOTE subset shape:", X_smote_enc.shape)

# 3) Apply SMOTE on the encoded subset (numeric, smaller, fits in RAM)
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_smote_enc, y_smote)

print("\nBefore SMOTE (subset):")
print(y_smote.value_counts())
print("\nAfter SMOTE (subset balanced):")
print(y_train_balanced.value_counts())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    precision_recall_curve,
)

def evaluate_model(name, model, X_val, y_val, threshold=0.5, plot_curves=True):
    # Probabilities for class 1
    y_proba = model.predict_proba(X_val)[:, 1]
    # Hard predictions
    y_pred = (y_proba >= threshold).astype(int)

    print(f"\n=== {name} @ threshold={threshold:.3f} ===")
    print("ROC-AUC:", roc_auc_score(y_val, y_proba))
    print(classification_report(y_val, y_pred, digits=4))

    # Confusion matrix
    cm = confusion_matrix(y_val, y_pred)
    plt.figure(figsize=(4,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} – Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    if plot_curves:
        prec, rec, _ = precision_recall_curve(y_val, y_proba)
        plt.figure(figsize=(5,4))
        plt.plot(rec, prec, label=name)
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.title(f"{name} – Precision–Recall Curve")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    return y_proba
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

log_reg = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        max_iter=1000,
        n_jobs=-1
    ))
])

log_reg.fit(X_train, y_train)
log_reg_proba_val = evaluate_model("Logistic Regression", log_reg, X_val, y_val)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_leaf=3,
    n_jobs=-1,
    random_state=42,
    class_weight=None   # dataset already balanced
)

rf.fit(X_train, y_train)
rf_proba_val = evaluate_model("Random Forest", rf, X_val, y_val)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

thr_list = np.linspace(0.1, 0.9, 30)
rows = []

for t in thr_list:
    preds = (rf_proba_val >= t).astype(int)
    rows.append((
        t,
        precision_score(y_val, preds, zero_division=0),
        recall_score(y_val, preds, zero_division=0),
        f1_score(y_val, preds, zero_division=0),
    ))

thr_df = pd.DataFrame(rows, columns=["threshold", "precision", "recall", "f1"])
thr_df.sort_values("f1", ascending=False).head(10)
```

```python
best_t_rf = thr_df.loc[thr_df["f1"].idxmax(), "threshold"]
best_t_rf
```

```python
rf_proba_val = evaluate_model("Random Forest (F1-opt)", rf, X_val, y_val, threshold=best_t_rf)
```

```python
from sklearn.neural_network import MLPClassifier

max_rows_mlp = 200_000  # adjust if it’s too slow/fast

if len(X_train) > max_rows_mlp:
    sample_idx = X_train.sample(n=max_rows_mlp, random_state=42).index
    X_train_mlp = X_train.loc[sample_idx]
    y_train_mlp = y_train.loc[sample_idx]
else:
    X_train_mlp = X_train
    y_train_mlp = y_train

mlp = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", MLPClassifier(
        hidden_layer_sizes=(128, 64),
        activation="relu",
        solver="adam",
        max_iter=20,
        learning_rate_init=1e-3,
        early_stopping=True,
        n_iter_no_change=3,
        random_state=42
    ))
])

mlp.fit(X_train_mlp, y_train_mlp)
mlp_proba_val = evaluate_model("MLP (ANN)", mlp, X_val, y_val)
```

## Final Model Selection: XGBoost

After comparing all candidate models on the validation set (Logistic Regression, Random Forest, ANN/MLP, and XGBoost), we selected **XGBoost** as the final model for deployment and test evaluation.

The choice is based on the following observations:

- **Highest discriminative performance:**  
  XGBoost obtained the **best ROC–AUC (~0.98)**, meaning it ranks clients by credit eligibility more accurately than Logistic Regression (~0.82), Random Forest (~0.95), and ANN (~0.93).

- **Best F1 score (balanced precision–recall):**  
  After tuning the decision threshold on the validation set, XGBoost achieved the **highest F1 score (~0.92)**, improving both precision and recall on the positive class compared to the other models. This is important because we want to detect as many eligible clients as possible while keeping false approvals under control.

- **Favorable confusion matrix:**  
  The confusion matrix for XGBoost shows **fewer false negatives and false positives** than the alternatives at their optimal thresholds, indicating a better trade-off between rejecting good clients and accepting bad ones.

- **Robustness on tabular data:**  
  Gradient-boosted trees are known to perform very well on large, structured/tabular datasets like ours. XGBoost captures non-linear relationships and feature interactions that simpler models (e.g., Logistic Regression) cannot, while remaining more efficient and easier to tune than deep neural networks for this problem.

For these reasons, **XGBoost with the F1-optimized threshold** is chosen as the final model. In the next cell, we evaluate this model once on the held-out **test set** to estimate its generalization performance.


```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

# Probabilities on test set
y_test_proba_xgb = xgb.predict_proba(X_test)[:, 1]

# Hard predictions using the same F1-opt threshold from validation
y_test_pred_xgb = (y_test_proba_xgb >= best_t_xgb).astype(int)

print("=== XGBoost – TEST SET (threshold = {:.3f}) ===".format(best_t_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_test_proba_xgb))
print(classification_report(y_test, y_test_pred_xgb, digits=4))

cm_test = confusion_matrix(y_test, y_test_pred_xgb)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(4,4))
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues")
plt.title("XGBoost – Confusion Matrix (Test)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
